## The Simulator Architecture

Before we investigate the C language in more detail, it is useful to first understand the general premise of how the Webots simulation operates.  The below image illustrates the `architecture` of the simulator:

<p align="center">
<br>
<img src="https://github.com/paulodowd/EMATM0053_21_22/blob/main/images/Webots_SimulatorController.png?raw=true">
<br>
</p>

In the above, we can conceptually seperate the operation of the simulator versus the robot controller.  Because we are simulating a process over time, it is necessary to "step" (advance) the simulator `discretely` (at fixed intervals).  When the simulator is stepped, the sensors are updated for the new environment status, and the actuators are provided their new instructions.  


<p align="center">
<br>
<img src="https://github.com/paulodowd/EMATM0053_21_22/blob/main/images/Webots_UpdateCycle.png?raw=true">
<br>
</p>

Whilst it is possible to step the simulator forwards within any part of your code (as illustrated below), it is extremely likely you will develop bugs and erroneous behaviours in your simulator robot.  This is because time will be able to advance from any part of your program, and there will be a loss of any general synchronisation to your program.  

<p align="center">
<br>
<img src="https://github.com/paulodowd/EMATM0053_21_22/blob/main/images/Webots_WrongSimStep.png?raw=true">
<br>
</p>

Instead, it is better to write a controller to not interfere with how time is incremented  - similar to the way real time advances regardless of our human actions.  Therefore, you should write your controller so that the `loop()` function returns at every time interval, and for your controller to re-evaluate present circumstances at every time interval.  


This is not especially convenient, however, it is much closer to how we would develop code for an embedded computing platform such as a microcontroller.

```c
/*********************************************
 Global variables.  
   Declare any global variables here to keep
   your code tidy.
**********************************************/

// simulation time step is 32ms
#define TIME_STEP 32  

// 3 IR ground sensors
#define NB_GROUND_SENS 3
#define GS_LEFT 0
#define GS_CENTER 1
#define GS_RIGHT 2
WbDeviceTag gs[NB_GROUND_SENS]; /* ground sensors */
unsigned short gs_value[NB_GROUND_SENS] = {0, 0, 0};

// Motors
WbDeviceTag left_motor, right_motor;

// LEDs
#define NB_LEDS 8
WbDeviceTag led[NB_LEDS];

// Proximity Sensors
#define NB_PS 8
WbDeviceTag distance_sensor[NB_PS];

// Position Sensors (encoders)
WbDeviceTag left_position_sensor, right_position_sensor;

```

#### Global Variables: #define
To look more closely:
```c
#define TIME_STEP 32
```
The above is called a `#define` ("hash-define"), and these can be thought of as a `find-and-replace` operator in your code. If your code has a data item which is used in many places and does not change, a `#define` is a good idea to use.  A `#define` is then referenced elsewhere in code by the reference (in this case `TIME_STEP`), and the compiler will substitute in the value (in this case, `32`).  

We can see this `find-and-replace` immediately in effect with the lines:
```c
#define NB_GROUND_SENS 3
...
unsigned short gs_value[NB_GROUND_SENS] = {0, 0, 0};
```
The above code creates a space in computer memory to store 3 `unsigned short` variables.  The value `3` comes from the `#define NB_GROUND_SENS 3` line of code.  The reference `NB_GOUND_SENS` was chosen to briefly represent "**N**um**b**er of **ground** **sens**ors".  The 3 memory elements named `gs_value` are then assigned the initial values of 0 (` = { 0, 0, 0 }`).  We will investigate `variables` in more depth soon.

`#define` are a good idea to use because: 
- they allow you to set constants in a convenient location in your program.  Note that, this means we **would not** use a `#define` for a value we expect to change within the program operation.
- they help to reduce errors and bugs, because you do not need to find all the values throughout your code - you can just change the `#define` value at the top of your program, meaning you only need to change something in a single location (reducing human error).
- they can be used to make your code more human-readable, also helping to reduce errors and bugs.
<p><br></p>


#### Global Variables: Special Cases

We can see several lines that look similar to:

```c
WbDeviceTag gs[NB_GROUND_SENS]; 
WbDeviceTag left_motor, right_motor;
WbDeviceTag distance_sensor[NB_PS];
WbDeviceTag left_position_sensor, right_position_sensor;
...
```
Hopefully you can notice a pattern - they all start with `Wb`.  Throughout the source code, you will notice highlighted code (yellow within Webots) that starts with `Wb` or `Wb_`.  These elements of code are keywords or functions relating uniquely to **W**e**b**ots.    These are all non-standard parts of the C programming language.  

The above four lines are all creating a reference (a name to access) the robot devices (the ground sensor, the motors, the distance sensors, the motor position encoders - respectively).  In these cases, the text starting with `Wb...` is the keyword to the compiler, and the next string of characters (e.g. left_motor) becomes the reference (name) we will use in the code.  

In C, you will see this format repeated often.  The format in question is:
```c
WbDeviceTag  left_motor;
<type>       <reference>
```

For example, we will soon come across many types of `variable`, which all take up different amounts of computer memory.  When these are defined in code, they will look something like:

```c
float  my_float_variable; // e.g. 0.001 or -99.9954
int    my_int_variable;   // e.g. -5 or 255
char   my_char_variable;  // e.g.  a or e
bool   my_bool_variable;  // e.g. false or true
<type>  <refrence>
```
In the above, we can see the same pattern - first the type is declared (float, int, char, bool) which indicates how much memory is to be allocated.  And then the reference is declared (the name assigned) so that elsewhere in code we can assign or read a value from that allocated memory.  Imagine - if we didn't create a reference (name), we would never be able to find the allocated memory again; and if we didn't declare the type, the compiler wouldn't know how much memory to set aside (all of it?  just one byte?).  

These `Wb` keyword references are later initialised (configured) in the `setup()` function, which we will look at later.  These `Wb` keywords and functions are the only non-standard parts of C you should encounter.

#### Function Prototypes

```c
/*********************************************
 Function prototypes.  
   Tells the C compiler what functions we have 
   written within this source file
**********************************************/

// One time setup, at beginning of simulation
void setup( );

// Controller to be called again and again.
void loop( );

// An example function that doesn't do anything
float doNothing( float f );
```

The above code indicates to the compiler which additional functions have been written beyond the `#include` libraries.  Creating function prototypes is also useful for the readibility of your code - it makes it clear what can be achieved with your code at the top of the source file, a bit like a contents page.

The function prototype has a similar format to the variable declaration we reviewed just prior.   Notice:
```c
// One time setup, at beginning of simulation
void   setup();
<type> <name>
```
The above line of code first has a type, and then a name which will we use to call the function (operate or execute it).  The functions `setup()` and `loop()` both have the type `void` prefixed in their declaration - this means that the function will not return any data as a result of it's operation.  

```c
// An example function that doesn't do anything
float doNothing( float f );
```
The above function prototype has a return type of `float`.  This means that when this function is called, it will return a value back.  This can be used to assign a the result of a function back into a variable.

```c
// An example of assigning the result of
// a function call.
float a;
a = doNothing( 10 );

```

The example function `doNothing(float f)` also has a declaration inside the parentheses (` float f`).  When we call a function, we often want it to do something or process some provided data.  When we pass some data to a function these are called `arguments`.  It is possible to have no arguments, such as the `setup()` and `loop()` examples.  It is also possible to have more than one argument given to a function.

```c
float x;
float c;
x = 10;
c = addTwoNumbers( x, x );

// A very simple function example.  Takes two float
// values as input, and returns their sum.
float addTwoNumbers( float a, float b ) {
  float result;
  
  // Functions should be used to encapsulate common
  // operations in your code.  Here, it is the sum 
  // of two numbers.
  result = a + b;


  // For examples sake, this assignment would not
  // change the original value of x.  Arguments 
  // provide only the value, not access to the 
  // variable.  Arguments can be treated as 
  // temporary variables.
  a = 20;

  // Pass back the value of result.
  return result;
}

```

It is important to know that the `function prototype` at the top of your source file must match the actual function declaration else where in your code.  For example:

```c
// An example function that doesn't do anything
void doNothing( int f );

// The function declaration
void doNothing( float f ) {
  return;
}
```
The above code will not compile, because the arguments have different types declared (`float` vs. `int`).  Common compile errors with functions are:
- if the `argument types` are different
- if the `return types` are different
- if the `functions names` are spelt differently
- if the function declaration or prototype is repeated more than once.

By now, you will likely feel like there is a lot to remember.  With practice, you will encounter common errors and learn to solve them quickly.  In fact, the errors you receive from the compiler should be regarded as a helpful friend! 

#### Main Thread

In this section we will review the `Main thread`.  The main thread is the heart of every `C` program.  The main thread is standard, and you should expect to always find a main thread within a C program.  

To explain the main thread as plainly as possible: every program that we run on a computer needs an `entry point`.  That is, when we ask our computer (a PC, Mac, microcontroller, microprocessor, etc) to run the program, what is the first part of the program to be run?  In C, we expect this to be the `main` function.  From the main function, our code can then call other functions, iterate (loop), call other programs, and many other operations.  Generally speaking, if the program finishes without any errors, the main function will return the value `0`, which is recognised as "success".

For our work within  Webots, we will be able to leave the main function exactly as it is presented in the source file and this labsheet.  During your working, it is expected that you'll develop further code to run from the `setup()` and `loop()` functions, which we will review soon.  In which case, the main function serves the following purpose, and does not need any changes:
- to provide an entry point to begin the simulation
- to call our `setup()` function, to prepare our robot controller
- to iteratively call our `loop()` function:
  - to update our robot controller, and send robot commands
  - to stop iterating when the simulation is ended.
- to provide an exit point, to finish the simulation

It is important to be aware that the main thread declaration can take different forms, and in this example it is nearly as simple as it can be:

```c
/*********************************************
 Main Thread. 
   This is the main thread of our C program.
   For this unit of study, it is recommended
   you do not alter the structure of main()
   below.  Instead, you should work from the
   functions setup() and loop().  You are 
   free to create your own useful functions.
**********************************************/
int main(void) {

  // Initialise Webots - must be done!
  wb_robot_init();

  // Code called here happens before the 
  // simulator begins running properly.
  setup();

  // Run the simulator forever.
  while( wb_robot_step(TIME_STEP) != -1 ) {
      
    // Call robot controller loop
    loop();
  }

  // Ask Webots to stop nicely
  wb_robot_cleanup();

  // Done.
  return 0;
}
```

If you are quick, the first thing you might notice is that the `main` function is created without a prototype earlier in our source file. In contrast, a loop operator within `countToTen()` (such as the example below) would have blocked `loop()` from running for a (very) short duration:  

```c
// An example of a blocking function
bool countToTen() {
  int i;
  i = 0;

  // microcontroller will be repeating
  // this 10 times
  while( i < 10 ) {
    i = i + 1;
  }
  return true;

}
```

In general, the example of counting to 10 would not be of significant detriment to the system.  However, hardware access or iteration over large arrays can be a problem.
